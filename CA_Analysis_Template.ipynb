```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random as rdm
```

```python
# Read Excel data as DataFrame, then convert to numpy matrix
# Make sure this script and your excel file are in the same directory!
Chronoamp_Data = pd.read_excel("example_data.xlsx", header=None, skiprows=2).to_numpy()
```

```python
## Clean raw data

# Set concentrations
num_trials = 3
Concentration_range_mM = np.array([0, 0.05, 0.10, 0.20, 0.50, 0.75, 1])
num_per_trial = len(Concentration_range_mM)

# Repeat concentration vector for all trials
C_mM = np.tile(Concentration_range_mM, num_trials)
C0 = C_mM / 1_000_000  # Convert mM to mol/mL

# Extract time and current data
time = Chronoamp_Data[:, 0]  # time in seconds
time = time.astype(float)    # ensure it's float, not object
# Extract every second column starting from col 1
Current_full = Chronoamp_Data[:, 1::2]  # uA
```

```python
# Plot Chronoamp Data (uA vs time) 
# This should recreate the PSTrace plot!

plt.figure(figsize=(8, 6))

# Generate a colorblind-accessible palette
num_lines = Current_full.shape[1]
palette = sns.color_palette("viridis", num_lines)
rdm.shuffle(palette)  # Shuffle colors for contrast

for i in range(num_lines):
    trial_num = i // num_per_trial + 1
    label = f"{float(C_mM[i]):.3g} mM, Trial {trial_num}"
    plt.plot(time, Current_full[:, i], color=palette[i], label=label)

plt.xlabel("Time (s)")
plt.ylabel(r"Current ($\mu$A)")
plt.title("Chronoamperometry of *Substance* at Selected Concentrations")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
## Plot with trials Averaged (uA vs time) 
# This plot should be included in offical reports/journal articles!

# Average the three trials for each concentration
Current_Uniques = np.zeros((Current_full.shape[0], num_per_trial))
for j in range(num_per_trial):
    for i in range(Current_full.shape[0]):
        # Select entries for concentration j every num_per_trial columns
        relevant_entries = Current_full[i, j::num_per_trial]
        Current_Uniques[i, j] = relevant_entries.mean()

# Plot averaged data
plt.figure(figsize=(8, 6))
num_lines = Current_Uniques.shape[1]
palette = sns.color_palette("viridis", num_lines)
np.random.shuffle(palette)

for i in range(num_lines):
    plt.plot(time, Current_Uniques[:, i], color=palette[i], label=f"{C_mM[i]} mM")

plt.ylim([-45, 5])
plt.xlabel("Time (s)")
plt.ylabel(r"Current ($\mu$A)")
plt.title("Chronoamperometry of *Substance* at Selected Concentrations (Averaged Trials)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
## Calibration Curve
# This plot should be included in offical reports/journal articles!

start_second = 100
end_second = 110
start_idx = start_second * 10  # zero-based indexing, 0.1s step => 10 samples per second
end_idx = end_second * 10

Current_avgs = np.zeros(Current_full.shape[1])

for i in range(Current_full.shape[1]):
    relevant_currents = Current_full[start_idx:end_idx + 1, i]
    Current_avgs[i] = relevant_currents.mean()

Current_avgs_A = Current_avgs / 1_000_000  # Convert uA to A

# Fit calibration curve
p = np.polyfit(C0, Current_avgs_A, 1)  # linear fit coefficients
y_fit = np.polyval(p, C0)

plt.scatter(C0, Current_avgs_A, color='k', s=20, label='Data Points', edgecolors='k')
plt.plot(C0, y_fit, 'r-', linewidth=2, label='Best Fit Line')

plt.xlabel('Concentration of Analyte (mol/mL)')
plt.ylabel('Average Current (A)')
plt.title('*Substance* Calibration Curve')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

print(f"The experimentally derived Slope is {p[0]}") # Slope proportional to sensitivity
```
